```python
def generate_annotations(file_name):
    # creating text from pdf
    pdf_file = f"{file_name}.pdf"
    doc = fitz.open(pdf_file)
    txt_total = ""
    for page_num, page in enumerate(doc.pages(), start=1):
        text = page.get_text()
        txt_total += text;

    # creating the text file
    with open('extracted_text.txt', 'w') as f:
        f.write(txt_total)

    # generating annotations with spacy
    nlp = spacy.load("en_core_web_lg")
    doc = nlp(txt_total)

    annotations = []

    named_entities = [(entity.text, entity.label_) for entity in doc.ents]
    for text,label in named_entities:
        annotations.append((text, label))


    # creating annotations dict
    classes = ['PERSON','ORG','GPE','LOC','DATE','TIME','MONEY','PERCENT','CARDINAL','QUANTITY','ORDINAL', 'FAC']

    annotations = {
        "classes" : classes,
        "annotations" : annotations
    }

    # dumping annotations as json
    json_object = json.dumps(annotations)

    with open("annotations.json", "w") as outfile:
        json.dump(json_object, outfile)

    return annotations
```
